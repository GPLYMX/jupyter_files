```python
import numpy as np
import os
from PIL import Image, ImageDraw
import torch
import cv2
def combine_img(folder_path):
    # 获取文件夹中的所有图像文件名
    image_files = [filename for filename in os.listdir(folder_path) if filename.endswith(".png")]
    # 加载灰度图像并添加到列表中
    image_files.sort()
    image_list = []
    for idx, img_path in enumerate(image_files):
        img = Image.open(os.path.join(folder_path, img_path)).convert("L")
        img = np.array(img)
        image_list.append(img)
        print(img.shape)

    image_list = np.array(image_list)
    image_list = image_list.transpose(1, 2, 0)
    print(image_list.shape)
    np.save('a.npy', image_list)
    return image_list
```

```python
import contextlib
import hashlib
import json
import os
import subprocess
import time
import zipfile
from multiprocessing.pool import ThreadPool
from pathlib import Path
from tarfile import is_tarfile
import glob

import cv2
import numpy as np
from PIL import ExifTags, Image, ImageOps
from tqdm import tqdm
def get_img_files(img_path):
    """Read image files."""
    prefix=''
    IMG_FORMATS = 'npy'
    HELP_URL = 'See https://docs.ultralytics.com/yolov5/tutorials/train_custom_data'
    fraction=1.0
    try:
        f = []  # image files
        for p in img_path if isinstance(img_path, list) else [img_path]:
            p = Path(p)  # os-agnostic
            if p.is_dir():  # dir
                f += glob.glob(str(p / '**' / '*.*'), recursive=True)
                # F = list(p.rglob('*.*'))  # pathlib
            elif p.is_file():  # file
                with open(p) as t:
                    t = t.read().strip().splitlines()
                    parent = str(p.parent) + os.sep
                    f += [x.replace('./', parent) if x.startswith('./') else x for x in t]  # local to global path
                    # F += [p.parent / x.lstrip(os.sep) for x in t]  # local to global path (pathlib)
            else:
                raise FileNotFoundError(f'{prefix}{p} does not exist')
        im_files = sorted(x.replace('/', os.sep) for x in f if x.split('.')[-1].lower() in IMG_FORMATS)
        # self.img_files = sorted([x for x in f if x.suffix[1:].lower() in IMG_FORMATS])  # pathlib
        assert im_files, f'{prefix}No images found'
    except Exception as e:
        raise FileNotFoundError(f'{prefix}Error loading data from {img_path}\n{HELP_URL}') from e
    if fraction < 1:
        im_files = im_files[:round(len(im_files) * fraction)]
    return im_files
```
